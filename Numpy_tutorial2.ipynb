```python
import numpy as np
```

# Sort array

```python
arr = np.array([2,1,5,3,7,4,6,8])
```

```python
# Sort the numbers in ascending order
np.sort(arr)
```

```python
# argsort -> Indirect sort along a specified axis
arr = np.array([10,30,20])
np.argsort(arr) # return indexes of the elements before sorting the array and place the index at the elements place after sorting
```

```python
# lexsort -> Indirect stable sort on mulitple keys
names = np.array(['A', 'B', 'C', 'D'])
age = np.array([21, 20, 21, 20])
marks = np.array([80, 90, 70, 95])

# Sort by age first, then marks i.e, np.lexsort((secondary_key, primary_key))
index = np.lexsort((marks, age))

print(index)
print(names[index])
```

```python
# searchsorted -> Find elements in a sorted array
arr = np.array([10, 20, 30, 40, 50])

print(np.searchsorted(arr, 25))
```

```python
# searchsorted -> Find multiple elements in a sorted array, if element is not there then return the index where that value will fall
print(np.searchsorted(arr, [20, 40, 30, 100, 70, 25, 5]))
```

```python
# searchsorted -> Find multiple elements as same in a sorted array
arr = np.array([10, 20, 20, 20, 30])

print(np.searchsorted(arr, 20, side='left'))
print(np.searchsorted(arr, 20, side='right'))
```

```python
# partition -> partial sort
arr = np.array([9, 1, 8, 2, 7, 3, 6, 4, 5])

print(np.partition(arr, 4)) # first 4 elements are sorted
```

```python
# partition -> partial sort
arr = np.array([9, 1, 8, 2, 7, 3, 6, 4, 5])
k = 3

print('Array elements until that indexed is printed', arr[:k])
print('Array elements until that index is printed in sorted form', np.partition(arr, 3)[:k])
```

# Concatenate Array

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])

np.concatenate((a,b))
```

```python
# axis means vertical stacking on top of each other
x = np.array([[1,2],[4,5]])
y = np.array([[5,9]])

print('Shape of x :', x.shape)
print('Shape of y:', y.shape)
np.concatenate((x,y), axis=0)
```

# Reshaping Arrays

```python
# Gives a new shape to an array without changing the data
# array you want to reproduce needs to have the same number of elemets as the original array
#(start, stop, step)
a = np.arange(2, 12, 2)
print(a)
print('Shape of a:', a.shape)
```

```python
# reshape the array to an array with 3 rows and 2 columns
b = a.reshape(3,2)
print(b)
```

```python
arr = np.arange(1,7)
print(arr)

#reshape to 3 rows and 3 columns
arr.reshape(3,2)
```

```python
#reshape to 2 rows and 2 columns
arr.reshape(2,3)
```
