```python
a = np.array(23)
print('No of Dimensions - ', a.ndim) # The number of dimensions of an array is contained in the ndim attribute.
print('Shape -', a.shape) # The shape of an array is a tuple of non-negative integers that specify the number of elements along each dimension.
print('Len of Shape - ', len(a.shape))
print('size -',a.size) # total number of elements in array is contained in the size attribute.
print('datatype -',a.dtype)
```

# How to create a basic array

This section covers np.zeros(),np.ones(),np.empty(),np.arange(),np.linspace()

```python
#Creating an array filled with 0's
np.zeros(2)
```

```python
np.zeros((2,3))
```

```python
np.zeros((2,3,4,3))
```

```python
#creating an array filled with 1's
np.ones(2)
```

```python
np.ones((2,3))
```

```python
np.ones((1,2,3,7))
```

```python
# Create an array and you want to override all the array elements later, So right now assign some garbage value to it Hence, create empty array
np.empty(3)
```

```python
np.empty((4,3))
```

```python
#Create an array with range elements
np.arange(7)
```

```python
#create an array with the range of evenly spaced intervals. (first number, last number, step size)
# Where the last number is not inclusive
np.arange(6,30,3)
```

```python
#Create an array with values that are spaced lineraly in a specified interval (first number, last number, number of elements)
np.linspace(0, 10, num=5)
```

```python
np.linspace(0,10,10)
```
