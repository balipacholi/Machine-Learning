# Convert a 1D array into a 2D array (how to add a new axis to an array)

```python
#np.newaxis and np.expand_dims to increase the dimensions of your existing array
#np.newaxis increases the dimensions of your array by one dimension when used once. eg-> 1D to 2D, 2D to 3D and so on.

a = np.array([1,2,3,4,5,6])
a.shape
print("ndim",a.ndim)

```

```python
#np.newaxis to add a new axis
row_vector = a[np.newaxis, :] # convert 1D to row vector
print("row_vector :", row_vector)
print ("shape :",row_vector.shape)
print("ndim",row_vector.ndim)
```

```python
col_vector = a[:, np.newaxis]
print("col_vector :", col_vector)
print ("shape :",col_vector.shape)
print("ndim",col_vector.ndim) # To print dimensions
```

```python
# use np.expand_dims to add an axis at index postion 1
col = np.expand_dims(a, axis=1) # column vector
print("col :", col)
print ("shape :",col.shape)
print("ndim",col.ndim)
```

```python
row = np.expand_dims(a, axis=0) #row vector
print("row :", row)
print ("shape :",row.shape)
print("ndim",row.ndim)
```

# Horizontal, vertical Stack Horizontal Split

```python
#Vertical Stack (vstack)
a1 = np.array([[1, 1],
              [2,2]])
a2 = np.array([[3,3],
              [4,4]])
np.vstack((a1,a2))
```

```python
#Split into several smaller arrays using hsplit
#specify the numbers of equally shaped arrays to return or the columns after which the division should occur
x = np.arange(1,25)
print("Original array\n",x)
x = x.reshape(2,12)
print("reshaped array\n", x)
#split this array into 3 equally shaped arrays
print("hsplit - array splitted into 3 equal parts\n")
np.hsplit(x,3)
```

# View & Copies

```python
# Create a new array object that looks at the same data as the original array(shallow copy)
# Numpy functions, as well as operations like indexing and slicing, will return views whenever possible.
# This saves memory and is faster (no cpoy of the data has to be made)
# Important :- Modifying the data in a view also modifies the original array!

arr = np.arange(6)
print("Value of arr\n", arr)
b = arr  #copying of arr to b array
print("Value of b\n", b)
# both arr and b will have same elements in array

# Updating a single element in b
b[4] = 10
print("Value of arr\n", arr) # But, the update will change the elements in the original array as well, Because both are shallow copy (pointing to same memory)
print("Value of b\n", b)
```

```python
# Slicing of array
arr = np.arange(10)
print("Value of arr\n", arr)
b = arr[2:] # B is sliced version of array
print("Value of b\n", b)

# Updating a single element value in b
b[3] = 70
print("Value of arr\n", arr) # It changes 3rd element of b i.e 5 to 70
print("Value of b\n", b)
```

```python
# Deep Copy - Make a complete copy of the array and its data
b2 = arr.copy()
print("Value of arr\n", arr)
print("Value of b2\n", b2)

# Updating b2 some elements
b2[1] = 40
b2[5] = 100
print("Value of arr\n", arr)
print("Value of b2\n", b2) # hence, in deep copy the elements that are changes will not be reflected in the original array as the memory address for both is different.
```
